#### Step-2(Data Preparation and Cleaning)

```python
#Displaying the Total Null Values in the dataset
df.isnull().sum()
```

```python
#Displaying the Total duplicate Values in the dataset
df.duplicated().sum()
```

```python
#Feature Engineering(Identifying and Seperating the input features and Target Variable)
X = df.drop('price',axis=1)
y = df['price']
print("The Shape of Features is:",X.shape)
print("The Shape of Target is:",y.shape)
```

#### Step-3(Model Building)

#One-hot Encoding - To Buid the Linear Regression we have to convert the Categorical Columns into 0/1 so we are converting the categorical columns by using One Hot Encoding

```python
c_col = ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']
X = pd.get_dummies(X,columns=c_col,drop_first=True)
```

```python
print("The Shape of Features After One-Hot Encoding:",X.shape)
```

```python
print("The New Column Names are:",X.columns)
```

```python
#Spliting the data into train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=45)
```

```python
#Building the Linear Regression Model
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

#### Step-4(Model Evaluation and Interpretation)

```python
#predicting the Model
y_pred = lm.predict(X_test)
y_pred
```

```python
#Evaluating the model using the r2_score and mean_absolute_error 
from sklearn.metrics import r2_score,mean_absolute_error
print("The r2_score of the model is:",round(r2_score(y_test,y_pred),3))
print("The mean_absolut_error of the model is:",round(mean_absolute_error(y_test,y_pred),3))
```
